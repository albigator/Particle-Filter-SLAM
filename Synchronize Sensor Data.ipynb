# Synchronizing Encoder 

```python
encoder.shape
```

```python
currCount=encoder[0,[0,1]]
encoder_sync=np.zeros((lidar.shape[0],3))
for i in range(lidar.shape[0]):
    index = np.argmax(time_encoder >= time_lidar[i])
    closestTime= time_encoder[index]
    encoder_sync[i,0]= closestTime #in encoder
    encoder_sync[i,[1,2]]= encoder[index,:]- currCount
    print(index, encoder_sync[i,[1,2]])
    currCount= encoder[index,:]
    if (i%(np.floor(lidar.shape[0]/10))==0):
        print("iteration #: ",i)
```

```python
print(encoder_sync)
```

```python
with open('encoder_sync.npy', 'wb') as f:
    np.save(f, encoder_sync)
with open('encoder_sync.npy', 'rb') as f:
    a = np.load(f)
```

# Synchronizing FOG

```python
deltaYaw=fog[:,2]
cumulativeYaw=np.zeros((fog.shape[0]))
cumulativeYaw[0]=deltaYaw[0]
for i in range(1,fog.shape[0]):
    cumulativeYaw[i]=deltaYaw[i]+cumulativeYaw[i-1]
    
cumulativeYaw.shape
print(cumulativeYaw)
print(cumulativeYaw.shape)
```

```python
for i in range(cumulativeYaw.shape[0]):
    if i%100==0:
        print(i, cumulativeYaw[i]*180/np.pi)
```

```python
pastYaw=0
```

```python
yaw_sync=np.zeros((lidar.shape[0],2))
cumulativeYaw_sync=np.zeros((115865))
for i in range(lidar.shape[0]):
    index = np.argmax(time_fog >= time_lidar[i])
    yaw_sync[i,0]= time_fog[index] #in encoder
    if(i==0):
        yaw_sync[0,1]= cumulativeYaw[index]
        pastYaw=cumulativeYaw[index]
        cumulativeYaw_sync[i]= cumulativeYaw[index]
    else:
        yaw_sync[i,1]= cumulativeYaw[index]- pastYaw
        pastYaw=cumulativeYaw[index]
        cumulativeYaw_sync[i]=cumulativeYaw[index]
        
    pastYaw=cumulativeYaw[index]
    if (i%(100)==0):
        print("iteration #: ",i, yaw_sync[i,1])
```

```python
print(yaw_sync.shape)
print(yaw_sync)
```

```python
for i in range(yaw_sync.shape[0]):
    if(i%100==0):
        print(i, yaw_sync[i,1]*180/np.pi)
```

```python
with open('cumulativeYaw_sync.npy', 'wb') as f:
    np.save(f, cumulativeYaw_sync)
with open('cumulativeYaw_sync.npy', 'rb') as f:
    myCumulativeYaw_sync = np.load(f)
    
print(myCumulativeYaw_sync.shape)
print(myCumulativeYaw_sync)
for i in range(myCumulativeYaw_sync.shape[0]):
    if i%100==0:
        print(i, myCumulativeYaw_sync[i]*180/np.pi)
```
